```python
# 剔除指定照片上除文字以外的景物
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

m1 = cv2.imread("h2.png",1)

m2 = cv2.subtract(m1[:,:,2], m1[:,:,0])  # 減掉藍
m3 = cv2.subtract(m2, m1[:,:,1])     # 減掉綠

m4 = cv2.bitwise_not(m3)  # 做色相相反

cv2_imshow(m1)
cv2_imshow(m4)
```
